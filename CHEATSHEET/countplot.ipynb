```python
import pandas as pd
import numpy as np
import ast
import matplotlib.pyplot as plt
import seaborn as sns

filepath = "../games.csv"
data = pd.read_csv(filepath,index_col = 0)

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the countplot with a color palette
sns.countplot(x='Rating', data=data, palette='viridis', hue='Rating', legend=True)

# Add a title and labels
plt.title('Count of Games by Rating', fontsize=20)
plt.xlabel('Rating', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

```python

# Create a new dataframe
df_genres = pd.DataFrame({
    'Genres': data['Genres'].tolist()
})

# Convert tuples back to lists
df_genres = df_genres.explode('Genres')

# Get the top 20 genres
top_genres = df_genres['Genres'].value_counts().nlargest(20).index

# Filter the dataframe to include only the top 20 genres
df_top_genres = df_genres[df_genres['Genres'].isin(top_genres)]

# Set the figure size
plt.figure(figsize=(14, 8))

# Create the countplot with a color palette
sns.countplot(x='Genres', data=df_top_genres, palette='hls', order=top_genres, hue='Genres', legend=True)

# Add a title and labels
plt.title('Count of Games by Top 20 Genres', fontsize=20)
plt.xlabel('Genres', fontsize=15)
plt.ylabel('Count', fontsize=15)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# Adjust the font size of x-axis labels for better readability
plt.xticks(fontsize=12)

# Show the plot
plt.show()
```
