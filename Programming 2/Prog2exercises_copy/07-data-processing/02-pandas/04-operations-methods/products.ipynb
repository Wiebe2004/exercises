```python
import pandas as pd

data = {
    'Product_ID': [101, 102, 103, 104, 105],
    'Product_Name': ['Laptop', 'Smartphone', 'Tablet', 'Headphones', 'Smartwatch'],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Electronics'],
    'Price': [1200, 800, 400, 100, 300],
    'Quantity_Sold': [50, 100, 80, 200, 120]
}

products = pd.DataFrame(data)
```

```python
print("Arithmetic operations")
print("Price * Quantity Sold:")
print(products['Price'] * products['Quantity_Sold'])
print()
print("Total revenue generated by all products:")
print((products['Price'] * products['Quantity_Sold']).sum())
```

```python
print("Staistical Aggregations")
print("Mean of Price & Quantity Sold:")
print(products[['Price','Quantity_Sold']].mean())
print()
print("Median of Price & Quantity Sold:")
print(products[['Price','Quantity_Sold']].median())
```

```python
print("Merging & Joining")
print(pd.concat([products, products], axis=0))
```

```python
print("Handling Missing Data")
print("Introduce missing values:")
products.loc[2, 'Price'] = pd.NA
products.loc[3,'Quantity_Sold'] = pd.NA
print(products)
```

```python
print("Fill missing values:")
products.loc[2, 'Price'] = pd.NA
products.loc[3,'Quantity_Sold'] = pd.NA
products_filled = products.fillna(products[['Price', 'Quantity_Sold']].mean())
print(products_filled)

#Alternative
# numeric_columns = ['Price', 'Quantity_Sold']
# products[numeric_columns] = products[numeric_columns].fillna(products[numeric_columns].mean())
# products
```

```python
print("Handling Missing Data")
print("Drop missing values:")
products_dropped = products.dropna()
print(products_dropped)
```
