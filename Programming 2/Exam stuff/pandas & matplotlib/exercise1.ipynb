How would you read a csv file? And how would you make use of it with pandas & matplotlib?

```python
csv_file = pd.read_csv("champions.csv")
```

Okay great! Now, let's do a few exercises. You'll get the hang of it by just practicing, reading & understanding what they all do.
It's prety much straightforward.

```python
champions = ['Riven', 'Renekton', 'Rumble', 'Katarina', 'Kassadin']
champions = csv_file[csv_file['Champion Name'].isin(champions)]
champions
```

How would you make a very simple graph?

```python
import numpy as np

# plt.plot(champions[champions.columns[0]], champions[champions.columns[5]])
# OR
champions.plot(x='Champion Name', y='Base Attack Damage', kind='scatter') # this makes more sense. You can also change the kind of plot you which to use.
plt.grid()
plt.show()
```

Great. Let's make it more detailed. Pandas & plot make it easier to show statistics of data.

```python
import matplotlib.patches as mpatches
import numpy as np

# colors = plt.cm.viridis(np.linspace(0, 1, len(champions)))
colors = ['purple', 'red', 'orange', 'green', 'blue']
champions.plot(x='Champion Name', y='Base Attack Damage', kind='bar', figsize=(6, 6), color=colors)

patches = [mpatches.Patch(color=colors[i], label=champion) for i, champion in enumerate(champions['Champion Name'])]

plt.title('Base Attack Damage of my Favorite Champions')
plt.xlabel('Champion Name')
plt.ylabel('Base Attack Damage')
plt.legend(handles=patches, loc='upper right')
plt.ylim([0, 100]) # limiting the height of the y-axis
plt.xticks(rotation='horizontal')  # make x-axis labels horizontal
plt.yticks(rotation='vertical')
plt.show()
```

Great, let's do a comparison now within that csv.

```python
mid = csv_file[csv_file['Role'] == 'Mid']
hp_500= mid[mid['Base Health'] < 500]
hp_500.plot(x='Champion Name', y='Base Health', kind='barh', figsize=(4, 4), width=.2)

for i, v in enumerate(hp_500['Base Health']):
    plt.annotate(str(v), (v, i))

plt.xlim([0, 600])
plt.legend(loc='upper right')
plt.show()
```
